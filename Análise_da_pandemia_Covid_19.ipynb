# **Covid-19 no Brasil - Perspectiva**

<center><img alt="Covid Background" width="50%" src="https://github.com/Diego875621/Covid_Data_Analysis/blob/main/Covid%20images/background.jpg?raw=true"> <center>

A doença do coronavirus 2019 (COVID-19) é uma doença contagiosa causa por um vírus, que como consequência, causa  a síndrome respiratória aguda severa (SARS), por isso, a cepa do virus que causa essa doença foi nomeado SARS-CoV-2. Primeiramente identificado na China, esse virus se espalhou por todo mundo, levando a uma pandemia.

Seus sintomas são variáveis, incluindo desde pacientes assintomáticos até  falta de ar severa, perda de olfato e apetite, tosse e febre. Foi relatado também até mesmo depressão e problemas neurológicos como sequelas dessa doença.

Várias medidas foram tomadas para tentar controlar sua rápida expansão. O distanciamento social, uso de máscaras e alcool em gel e também o lockdown global (restrição de sair de casa e fechamento de estabelecimentos/viagens) foram as principais maneiras preventivas antes do total desenvolvimento de vacinas.

Diante da gravidade da doença, os sistemas de informações e epidemiologia foram e são essenciais para observar e registrar, por exemplo, quais os principais locais afetados, qual população é mais vunerável, entre outros questionamentos importantes. 

Dessa forma, vários sistemas de dados governamentais e não-governamentais trabalharam com um foco só, analisar dados e tomar decisões a partir delas. Um dos principais foi o *Our World In Data*, que continua registrando praticamente em tempo real as várias mudanças no cenario da pandemia Covid-19. 

**Será desse repositório (Our World In Data) que esta análise terá como base, especificamente os dados relacionados ao Brasil.**





# Dicionário de variáveis

<center><img alt="Dictionary" width="50%" src="https://github.com/Diego875621/Covid_Data_Analysis/blob/main/Covid%20images/list.jpg?raw=true"></center> <center> 

Para ajudar nossa compreensão, será um informado um dicionário de variáveis.

iso_code - Código Reconhecido Internacionalmente formado por 3 letras que pode ser usado para ser referir a países

continent - Continente da localização geográfica

location - Localização geográfica

date - Data de observação

total_cases - Número de casos confirmados. Pode incluir casos prováveis

new_cases - Número de casos novos. Valores ausentes representam números negativos causados por correção de data

new_cases_smoothed - Novos casos confirmados, nivelados por 7 dias.

total_deaths - Mortes totais atribuídas ao Covid-19.

new_deaths - Novas mortes atribuídas ao Covid-19.

new_cases_smoothed - Novas mortes atribuídas ao Covid-19 (nivelados por 7 dias)

total_deaths_per_million - Número total de casos confirmados por 1000000 de pessoas

new_deaths_per_million - Novos casos confirmados por 1000000 de pessoas

new_deaths_smoothed_per_million - Novos casos confirmados por 1000000 de pessoas (nivelados por 7 dias)

reproduction_rate - Estimativa em tempo real da proporção de reprodução (R) do Covid-19

icu_patients - Número de paciente com Covid-19 em Unidades de Terapia Intensiva (UTI) em um determinado dia

icu_patients_per_million - Número de paciente com Covid-19 em Unidades de Terapia Intensiva (UTI) em um determinado dia por 1000000 de pessoas

hosp_patients - Número de pacientes com Covid-19 hospitalizados em determinado dia

hosp_patients_per_million - Número de pacientes com Covid-19 por 1000000 de pessoas

weekly_icu_admissions - Número de pacientes com Covid-19 admitidos para a UTI em determinada semana

weekly_icu_admissions_per_million - Número de pacientes com Covid-19 admitidos para a UTI em determinada semana por 1000000 de pessoas

weekly_hosp_admissions - Número de pacientes com Covid-19 admitidos em hospitais recentemente, em determinada semana

weekly_hosp_admissions_per_million - Número de pacientes com Covid-19 admitidos em hospitais recentemente determinada semana por 1000000 de pessoas

total_tests - Número total de testes para Covid-1

new_tests - Novos testes para Covid-19 (para dias consecutivos)

total_tests_per_thousand - Número total de testes para Covid-19 por 1000 pessoas

new_tests_per_thousand - Novos testes para Covid-19 por 1000 pessoas

new_tests_smoothed - Novos testes para Covid-19 (nivelados por 7 dias)

new_tests_smoothed_per_thousand - Novos testes para Covid-19 (nivelados por 7 dias) por 1000 pessoas

positive_rate - Proporção dos testes para Covid-19 que são positivos dentro de uma média de 7 dias (este é o inverso do tests_per_case)

tests_per_case - Testes conduzidos por novos casos confirmados de Covid-19 dentro de uma média de 7 dias (este é o inverso do positive_rate)

tests_units - Unidades de testes usadas pela localização para reportar suas informações sobre testes

total_vaccinations - Número total de doses de vacinas de Covid-19 administradas

people_vaccinated - Número total de pessoas que receberam pelo menos uma dose de vacina de Covid-19

people_fully_vaccinated - Número total de pessoas que receberam todas as doses de vacina por determinado protocolo de vacinação.

total_boosters - Número total de doses de reforço da vacina (doses administradas além do protocolo padrão de vacinação)

new_vaccinations - Novas vacinações (calculado por dia consecutivos)

new_vaccinations_smoothed - Novas vacinações (niveladas por 7 dias)

total_vaccinations_per_hundred - Número total de vacinações a cada 100 pessoas

people_vaccinated_per_hundred - Número total de pessoas que receberam pelo menos uma dose de vacina a cada 100 pessoas

people_fully_vaccinated_per_hundred - Número total de pessoas que receberam todas as doses de vacina por determinado protocolo de vacinação a cada 100 pessoas

total_boosters_per_hundred - Número total de doses de reforço administradas a cada 100 pessoas.

new_vaccinations_smoothed_per_million - Novas doses de vacina administradas (niveladas por 7 dias) por 1000000 pessoas.

new_people_vaccinated_smoothed - Número diário de pessoas recebendo a primeira dose da vacina (nivelado por 7 dias)

new_people_vaccinated_smoothed_per_hundred - Número diário de pessoas recebendo a primeira dose da vacina (nivelado por 7 dias) por 100 pessoas.

stringency_index - Índice de Estrigência de resposta governamental, é uma medida composta por 9 indicadores de respostas incluindo fechamento de escolas, fechamento de locais de trabalho e proibição de viagens, em escala de 0 a 100 (100 = resposta mais estrita)

population - População

population_density - Número de pessoas dividida pela área, medida em quilômetros quadrados

median_age - Idade média da população

aged_65_older - Corte da população com idade de 65 anos ou mais velha

aged_70_older - Corte da população com idade de 70 anos ou mais velha

gdp_per_capita - Produto Interno Bruto em paridade com poder de compra (em dólares internacionais)

extreme_poverty - Porção da população vivendo em extrema pobreza, data mais recente disponível: desde 2010

cardiovasc_death_rate - Proporção de morte por doenças cardiovasculares em 2017

diabetes_prevalence - Prevalência de diabates (% da população entre 20-79 anos de idade) em 2017

female_smokers - Proporção de mulheres fumantes

male_smokers - Proporção de homens fumantes

handwashing_facilities - Porção da população com instalação básica de lavagem de mãos nas instalações.

hospital_beds_per_thousand - Leitos hospitalares por 1000 pessoas, ano mais recente disponível desde 2010

life_expectancy - Expectativa de vida ao nascer em 2019

human_development_index - Índice de Dessenvolvimento Humano - índice que avalia os 3 níveis básicos de dessenvolvimento humano - vida longa e saudável, conhecimento e um decente padrão de vida, valores de 2019

excess_mortality - Diferença Percentual do número de mortes semanais ou mensais de 2020-2021 e do valor projetado para o mesmo período

excess_mortality_cumulative - Diferença acumulativa do número de mortes desde 1 de Janeiro de 2020 e do valor projetado acumulado para o mesmo período

excess_mortality_cumulative_absolute - Diferença acumulativa do número de mortes desde 1 de Janeiro de 2020 e do valor projetado para o mesmo período

excess_mortality_cumulative_per_million - Diferença acumulativa do número de mortes desde 1 de Janeiro de 2020 e do valor projetado acumulado para o mesmo período por um milhão de pessoas.

Podemos percerber que esse vírus não é discriminatório, afetando tanto pessoas de alta renda quanto pessoas de baixa a média renda. Isso deve tanto pelo vírus ser relativamente novo (quanto a questão de ser fatal para humanos) e também da nossa reação, pois não estávamos preparados para tamanho nível de contaminação mundial.

Porém isso ainda não é conclusivo, pois cada país tratava o Covid-19 de acordo com seu manejo. Há países que testaram mais pessoas, ou contaram com maior estrutura de hospitais e centros de triagem e assim, poderiam rapidamente ter o diagnóstico e assim atender/orientar o paciente da melhor maneira.

É possível observar que o vírus foi letal para vários páises, independente do Indíce de Dessenvolvimento Humano (IDH), porém, houve um pico do total de mortes nos países com IDH entre 0.7-0.8.

Como podemos ver abaixo, o avanço do número de casos totais do Covid-19 aumentou rapidamente ao longo do tempo, após o mês de outubro de 2020 podemos ver um acelerado crescimento e depois de Janeiro de 2022 ouve um pico de incidência e posteriormente atingindo um certo platô a partir de Abril de 2022. Esse rápido crescimento se deve à rápida proliferação do vírus, pois é um vírus facilmente transmitido por fala, tosse e espirros, e somente um espiro pode se extender por até mais de 2 metros e contaminar superfícies. O vírus pode ser carreado até mesmo por um aperto de mão. Além disso, mesmo com as campanhas de vacinação, certa parte da população se recusa a tomar vacinas.

Os lugares mais afetados, considerando a contagem até a data de 05-05-2022 com maior número de mortes foram Europa, América do Norte, Ásia, América do Sul e Índia, em ordem decrescente.

Já a estabilização do número de casos e mortes veio com o advento da vacinação em massa da população. Uma pessoa vacinada, mesmo se contaminando pelo Corona Vírus, tem uma chance muito menor de manifestar sintomas da doença do Covid-19 e mesmo aqueles que manifestam normalmente não requerem internação ou cuidados hospitalares.
E com as doses de reforço, os números podem se tornar cada vez menores, pois as doses de reforço tem a capacidade de aumentar ainda mais a proteção do corpo e "treinar" o nosso corpo para mais facilmente reconhecer e combater o vírus.

# **Covid-19 no cenário Brasileiro**

<center><img alt="Covid_mask" width="50%" src="https://github.com/Diego875621/Covid_Data_Analysis/blob/main/Covid%20images/surgical-mask-brasil.jpg?raw=true"></center> <center> 

No Brasil, a pandemia também deixou vários rastos de destruição assim como no mundo todo. Como nenhum país estava preparado para tamanha onda de contaminação, juntamente com alta taxa de hospitalizações e mortes, o Covid-19 sobrecarregou vários hospitais, e muitos pacientes que não puderam receber a devida atenção médica devida à superlotação.

A maioria dos estabelecimentos, escolas, festas e viagens foram todas fechados ou cancelados. Também foi decretado estado de Lockdown, isso é, além do fechamento da maioria das atividades, a população foi notificada a se manter em casa o máximo de tempo possível e sair somente caso necessário, como para comprar mantimentos ou medicamentos.


A economia mundial foi profundamente atingida pelas consequências da pandemia. Como não podíamos ir ao local de trabalho, o trabalho remoto em casa (também chamado Home Office) foi instalado em diversas empresas, porém poucos estavam acostumados com essa modalidade, e a adaptação levou um bom tempo. Ainda, há aquelas empresas que seu modelo não tinha como adaptar via remota, e com isso, muitos trabalhadores foram suspensos ou até mesmo demitidos, inclusive várias empresas não conseguiram sobreviver, o que levou a falência e destruição da economia para muitos brasileiros.

Além de abalar a economia, a pandemia afetou também a saúde mental de várias pessoas. A taxa de doenças psiquiátricas, como depressão e ansiedade, aumentou drasticamente. Com a falta de emprego e consequente falta de dinheiro para movimentar a economia de muitas famílias, o estresse se tornou constante, pois além da alta taxa de desemprego, muitos produtos aumentaram de preço por conta da escassez de produtos provocados pela baixa produção e alta procura.

# **Vacinação**
<center><img alt="Vaccine" width="50%" src="https://github.com/Diego875621/Covid_Data_Analysis/blob/main/Covid%20images/kids%20vaccine.jpg?raw=true"></center> <center> 

O início da vacinação representou um grande marco mundial, pois é a principal forma de proteção contra o Covid-19. A vacina funciona estimulando o nosso organismo com pedaços inativos do vírus, como por exemplo uma proteína específica, seu RNA ou um pedaço do seu envoltório plasmático. 

Essa porção do vírus é suficiente para estimular o sistema imune porém sem causar a doença, e com isso é gerado células imunes de memória que podem gerar anticorpos específicos que podem rapidamente inativar e inibir a proliferação do Covid-19 e impedir uma futura contaminação.

Várias vacinas foram sendo desenvolvidas ao longo do tempo. A CoronaVac (Butatan), a Covishield (Astrazeneca), Corminaty (Pfizer) e a Janssen Vaccine (Janssen-Cilag) foram as principais vacinas aplicadas no Brasil.
