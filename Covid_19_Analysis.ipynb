# **Covid-19 - Perspective**

<center><img alt="Covid Background" width="50%" src="https://github.com/Diego875621/Covid_Data_Analysis/blob/main/Covid%20images/background.jpg?raw=true"> <center>

<p align = justify> Coronavirus disease (COVID-19) is a contagious disease caused by a virus, which as a consequence causes severe acute respiratory syndrome (SARS), so the virus strain that cause this disease has been named SARS-CoV-2. First identified in China, this virus has spread worldwide, leading to a pandemic.

<p align = justify>Its symptoms are variable, ranging from asymptomatic patients to severe breath shortness, loss of smell and appetite, coughing, and fever. Even depression and neurological problems have been reported as sequelae of this disease. 

<p align = justify>Various measures have been taken trying to control the rapid spread of the virus. Social distacing, the use of mask and alcohol gels, and also global lockdown (restrictions on leaving house and closing down businesses/travel) were the main preventive measures before the full development of vaccines.

<p align = justify>Given the severity of the disease, information systems and epidemiology were and are essential to observer and record, for instance, which are the main affected locations, which population is more vulnerable, among other important questions.

<p align = justify>In this way, several governmental and non-governmental data systems worked with one focus - to analyze data and with it, information make better decisions. One important company was "Our World in Data", which continues to record the various COVID-19 scenario changes almost in real-time

**This analysis will be based on this repository Our World In Data and will be specificallt related to Brazil**

```python
!pip install missingno -q

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno

%matplotlib inline

from datetime import date
```

We can start by first checking how big is the dataset that we are working on and check it's number of variables and inputs.

```python
# Checking the Dataframe

df = pd.read_csv("https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv", parse_dates=True)
df.head()

# Checking variables and inputs

print('Dataset dimension', '\nVariables: ',df.shape[1],
      '\nInputs: ', df.shape[0])
```

```python
# Dataset columns
df.columns
```

<p align = justify>It is noticeable that our dataset has a large number of variables, with the incredible number of 67 variables. It is important to state that this dataset is constantly been updated. In order to simplify our understanding and analysis, we will focus on a few variables as key factors for this exploratory analysis: `people_vaccinated`, `total_deaths`, `total_cases`, `location` and `date`.   

<p align = justify>For example, after the Covid vaccinations started, it was possible to also record the number of people vaccinated(`people_vaccinated`) and the number of patients who got boosters (`total_boosters`).

<p align = justify>Realizing that the variable `date` is set to object, we must transform it to *datetime*.

```python
# Date to datetime
df.date = pd.to_datetime(df.date)
```

```python
# Variables types
df.dtypes
```

# **Missing data**

We can also check the amount of missing values (Na) from the dataset. From the matrices and the table generated, we can see that there are many missing values especially in the variable involving vaccination, patients in Intensive Care Units (ICU) and mortality.

Remembering again that we should also look for reasons for the missing data. Because, as it involves a global problem, the data is constantly being updated.

```python
# Missing values matrix
msno.matrix(df, labels=True);
```

```python
# NaN percentage
(df.isnull().sum() / df.shape[0]).sort_values(ascending=False) [0:20]
```

```python
# 'people_vaccinated' Missing values (Percentage)
(df['people_vaccinated'].isnull().sum() / df.shape[0])
```

```python
# Filling Na of 'people_vaccinated' with 0
df.people_vaccinated.fillna (0, inplace=True)

(df['people_vaccinated'].isnull().sum() / df.shape[0])
```

We can see the highest number of cases for the current date (05/18/2022) recorded has Europe at first place, followed by Asia, North America, and South America (the first two entries have been hiden from view because they do not specify the places, so the count of [2:12].

For the total number of deaths for the same date, we have Europe again in the first place, followed by North America, Asia and South America.

So, altough COVID-19 was at first spread over much of Asia, especially China, other places like Europe and North America were hard hit.

```python
# Current date total cases
df.loc[df.date == '2022-05-18'].sort_values(by="total_cases", ascending=False) [2:12]
```

```python
# Current date total deaths
df.loc[df.date == '2022-05-18'].sort_values(by="total_deaths", ascending=False) [3:10]
```

```python
# Places with more deaths
df_morte = df.loc[df.date == '2022-05-18', ['location', 'total_deaths']].sort_values(
    by="total_deaths", ascending=False) [2:12]
df_morte
```

```python
fig, ax = plt.subplots()

df_morte.plot(x='location', y='total_deaths', kind='barh', ax=ax);
ax.set_title("Total deaths x Localization")
ax.invert_yaxis()
plt.show()
```

<p align = justify>We can see that the this virus is not discriminatory, affecting both high income people and low to middle income people. This is due both the fact that the virus is relatively new (as to whether it is fatal for humans) and also to our reaction, because we were not prepared to face such high level of worldwide contaminations.

<p align = justify>But this is not yet conclusive, because each country treated COVID-19 according to its management. There are countries that have tested more people, or have more hospital and screening centers, and thus could quickly have the diagnosis and in that way, assist/guide the patient in the best way.

<p align = justify>Through the comparison of the total number of deaths in relation to the Gross Domestic Product (GDP) per capita, we can see that the highest number of deaths is in an area of less than 200000 of GDP per person. It can been as well that the virus was lethal in many countries, regardless of their Human Development Index (HDI), but the total number of deaths peaked in countries with an HDI between 0.7-0.8.

# **Covid-19 on the Brazilian scenario**

<p align = justify>
In Brazil, the pandemic also left several trails of destruction as well as worldwide. As no country was prepared for such waves of contamination, along with the rate of hospitalizations and deaths, COVID-19 overwhelmed several hospitals, and many patients were unable to receive proper medical attention due to overcrowding.
<p align = justify>
Most establishments, schools, events/parties, and trips were all closed or canceled - a state of Lockdown was declared. That is, in addition to the closure of most activities, the population was told to stay at home as long as possible and only go out if necessary, such as to buy groceries or medicine.
<p align = justify>
The world economy was deeply hit by the consequences of the pandemic. As we could not go to the workplace, Working from home was installed in several companies, but few were used to this modality, and the adaptation took a long time. Still, there are companies and segments that by their inherited way of work, working from home was very hard or even impossible. With this, many workers were suspended or even fired, and several companies did not survive, which led to bankruptcy and destruction of the economy for many Brazilians.
<p align = justify>
Besides shaking the economy, the pandemic also affected the mental health of many people. The rate of psychiatric illnesses, such as depression and anxiety raised significantly. With the lack of jobs and production and the consequent lack of money to move the economy for many families, stress has become constant, because as the high unemployment rate increased, many products have increased their price due to the shortage of products caused by low production and high demand.
<p align = justify>
Below we will take a look at the dataset, however, specific to Brazil.

```python
# Looking specifically at the dataset for Brazil
df_brazil = df.loc[df.location == "Brazil"].copy()
df_brazil
```

```python
print('Dataset Dimensions (Brazil)', '\nVariables: ',df_brazil.shape[1],
      '\nInputs: ', df_brazil.shape[0])
```

```python
# Correlation 'total_deaths' x 'total_cases'
corr2 = df_brazil.corr().loc["total_deaths", "total_cases"]
print(f'Correlation total_deaths x total_cases {corr2:.2f}')
```

<p align = justify>
From the graphs below we can see that the number of total cases and total deaths in Brazil follow a similar wordwide trend, which is corroborated by the correlation of 0.98 between the two variables. This indicates that there has probably been a high mortality rate linked to the number of cases. 
<p align = justify>
We can also see that from January 2022 we have a big increase in the number of cases and mortality as well, and after this period there is apparently a plateau, probably responsible for the mass vaccination.

```python
# Plot total cases x total deaths

x = df_brazil['date']
y = df_brazil['total_cases']
y2 = df_brazil['total_deaths']

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15,10))
ax1.set_title('Total cases in Brazil')
ax1.set_xlabel('Date')
ax1.set_ylabel('Cases')
ax1.plot(x, y)

ax2.set_title('Total cases in Brazil (Log)')
ax2.set_xlabel('Date')
ax2.set_ylabel('Cases')
ax2.set_yscale('log')
ax2.plot(x, y)

ax3.set_title('Total death in Brazil')
ax3.set_xlabel('Date')
ax3.set_ylabel('Death')
ax3.plot(x, y2)

ax4.set_title('Total death in Brazil (Logartimico)')
ax4.set_xlabel('Date')
ax4.set_ylabel('Death')
ax4.set_yscale('log')
ax4.plot(x, y2)
plt.tight_layout()
```

<p align = justify>
It is possible to find the difference between the first recorded case of COVID-19 and the first recorded death. To do this, as a hypothesis, let us consider that the missing values for  `new_deaths` and `new_cases` are filled with the value 0, i.e., no cases or deaths during those dates.

# **Vaccination**
<center><img alt="Vaccine" width="50%" src="https://github.com/Diego875621/Covid_Data_Analysis/blob/main/Covid%20images/kids%20vaccine.jpg?raw=true"></center> <center> 

<p align = justify>
The start of vaccination represented a major milestone worldwide, as it is the main form of protection against COVID-19. the vaccine works by stimulating our body with inactive pieces of the virus, such as a specific protein, its RNA, or a piece of its viral envelope.
<p align = justify>
This portion of the virus is enough to stimulate the immune system but without causing disease, and with this process, memory immune cells are created, and those can synthesize specific antibodies that can quickly inactivate and inhibit the proliferation of the virus and then prevent further contamination.
<p align = justify>
Several vaccines have been developed over time. CoronaVac (Butatan), Covishield (Astrazeneca), Comirnaty (Pfizer), and Janssen Vaccine (Janssen-Cilag) were the main vaccines applied in Brazil.

[Anvisa - Vacinas](<https://www.gov.br/anvisa/pt-br/assuntos/paf/coronavirus/vacinas>)

```python
df_brazil.loc[df_brazil["people_vaccinated"] > 0]
```

<p align = justify>
The initial date of vaccination in Brazil recorded from January 17, 2021 to the present day. By approximately April 2022 we can see a certain plateau in the number of people vaccinated. 
<p align = justify>
In the chart below, it is possible to observe the rapid adherence to vaccines by Brazilians. Approximately 1.83 * 10^9 Brazilians have taken at least one dose of one type of vaccine, and compared to the number of Brazilians, approximately 2.14 * 10^8, according to the Brazilian Institute of Geography and Statistics (IBGE), we have 85.36%, which corresponds to a high portion of the Brazilian population vaccinated.
<p align = justify>
The parts of the graph that seem to lower to 0 probably occur because there were missing values on those dates, and so as we filled in 0, it appears that way on the graph.

# **Conclusions**
<p align = justify>
According to this exploratory analysis of the COVID-19 from the "Our World in Data" repository, it can be stated that Europe, North America, and Asia were the top 3 locations most affected by both moratily and the highest number of cases.
